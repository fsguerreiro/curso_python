```python
import pandas as pd
from flask_ngrok import run_with_ngrok
from flask import jsonify, Flask
```

```python
d = [{"Number": 1, "Name": "Mahesh", "Age": 25, "City": "Bangalore", "Country": "India"},
     {"Number": 2, "Name": "Alex", "Age": 26, "City": "London", "Country": "UK"},
     {"Number": 3, "Name": "David", "Age": 27, "City": "San Francisco", "Country": "USA"},
     {"Number": 4, "Name": "John", "Age": 28, "City": "Toronto", "Country": "Canada"},
     {"Number": 5, "Name": "Chris", "Age": 29, "City": "Paris", "Country": "France"}]

app = Flask(__name__)

run_with_ngrok(app)

@app.route('/api/example', methods=['GET'])



@app.route("/") #code to assign HomePage URL in our app to function home.

def home():
    '''
    The entire line below must be written in a single line.
    '''
    return "<marquee><h3> TO CHECK INPUT ADD '/input' TO THE URL AND TO CHECK OUTPUT ADD '/index' TO THE URL. </h3></marquee>"


@app.route("/input") #code to assign Input URL in our app to function input.

def input():
    return jsonify(d) # "d" is the dictionary we defined


@app.route('/index', methods=['GET','POST']) #output page

def panda():
    tabela = pd.DataFrame.from_dict(d)
    tabela = tabela.to_html(col_space=100, header="true", index=False, justify='left', table_id="table")

    return tabela


if __name__ == '__main__':
    app.run()
```
