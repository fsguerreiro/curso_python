```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/Gapminder.csv", sep=";")
```

```python
df = df.rename(columns={"country": "Pais", "continent": "Continente", "year": "Ano", 
                   "lifeExp": "Exp. vida", "pop": "Populacao", "gdpPercap": "PIB per capita"})
df.head(10)
```

```python
# Checando quantos elementos nulos há no dataframe:
df.isnull().sum()
```

```python
# Verificar quantos paises não tem continente preenchido: 
df1 = df.loc[df["Continente"].isnull()]
len(pd.unique(df1["Pais"]))

```

```python
# Agrupando população total por ano:
df2 = df.groupby(["Ano"])["Populacao"].sum()
df2 = pd.DataFrame(df2).reset_index()
df2.head(10)
```

```python
fig = plt.figure(figsize =(10, 4))
plt.bar(df2["Ano"], df2["Populacao"])
plt.xlabel('Ano', fontweight ='bold', fontsize = 12)
plt.ylabel('População Mundial', fontweight ='bold', fontsize = 12)
plt.show()
```

Como verificado por gráfico, apenas os anos terminados em 7 e 2 possuem dados da população de todos os países.

```python
df27 = df2[df2['Ano'] % 5 == 2].reset_index(drop=True)
df27.head()
```

```python
fig = plt.figure(figsize =(10, 4))
plt.bar(df27["Ano"], df27["Populacao"], width=4)
plt.xlabel('Ano', fontweight ='bold', fontsize = 12)
plt.ylabel('População Mundial', fontweight ='bold', fontsize = 12)
plt.xticks(np.arange(1952, 2008, step=5))
plt.show()
```

```python
df3 = df.groupby(["Ano"])["Exp. vida"].mean()
df3 = pd.DataFrame(df3).reset_index()
df3.head(10)
```

```python
fig = plt.figure(figsize =(10, 4))
plt.bar(df3["Ano"], df3["Exp. vida"])
plt.xlabel('Ano', fontweight ='bold', fontsize = 12)
plt.ylabel('Expectativa de vida', fontweight ='bold', fontsize = 12)
plt.show()
```

```python
df272 = df3[df3['Ano'] % 5 == 2].reset_index(drop=True)
df272.head()
```

```python
fig = plt.figure(figsize =(10, 4))
plt.bar(df272["Ano"], df272["Exp. vida"], width=4)
plt.xlabel('Ano', fontweight ='bold', fontsize = 12)
plt.ylabel('Expectativa de vida', fontweight ='bold', fontsize = 12)
plt.xticks(np.arange(1952, 2008, step=5))
plt.show()
```

```python
fig = plt.figure(figsize =(10, 4))
plt.scatter(df["PIB per capita"], df["Exp. vida"])
plt.xlabel('PIB per capita', fontweight ='bold', fontsize = 12)
plt.ylabel('Expectativa de vida', fontweight ='bold', fontsize = 12)
# plt.xticks(np.arange(1952, 2008, step=5))
plt.show()
```

```python
df4 = df.groupby(["Ano"])["PIB per capita"].mean()
df4 = pd.DataFrame(df4).reset_index()
df273 = df4[df4['Ano'] % 5 == 2].reset_index(drop=True)
fig = plt.figure(figsize =(10, 4))
plt.bar(df273["Ano"], df273["PIB per capita"], width=4)
plt.xlabel('Ano', fontweight ='bold', fontsize = 12)
plt.ylabel('PIB per capita mundial', fontweight ='bold', fontsize = 12)
plt.xticks(np.arange(1952, 2008, step=5))
plt.show()
```
